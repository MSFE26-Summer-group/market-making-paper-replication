```python
# Standard imports — isort will enforce this order via nbqa
from __future__ import annotations

import duckdb
import pandas as pd
```

## Dataset

```python
subset = duckdb.query(
    """
    SELECT *
    FROM '../data/btc_usdt_20221019_20221030_lob.parquet'
    LIMIT 10
"""
).df()
```

```python
subset.head()
```

```python
column_names = subset.columns.tolist()
print(f"The length of the column names is: {len(column_names)}")
```

### Checking which columns are empty

```python
for col in column_names:
    if subset[col].isna().all():
        print(f"Column '{col}' is empty.")
```

### Check how many columns have are there for ask and bid

```python
ask_price_cols = []
bid_price_cols = []
ask_size_cols = []
bid_size_cols = []

for col in column_names:
    if ".ask_price" in col:
        ask_price_cols.append(col)
    elif ".bid_price" in col:
        bid_price_cols.append(col)
    elif ".ask_size" in col:
        ask_size_cols.append(col)
    elif ".bid_size" in col:
        bid_size_cols.append(col)

print(f"Number of Ask price columns: {len(ask_price_cols)}")
print(f"Number of Bid price columns: {len(bid_price_cols)}")
print(f"Number of Ask size columns: {len(ask_size_cols)}")
print(f"Number of Bid size columns: {len(bid_size_cols)}")
```
